```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
try:
    c = a/b
except ZeroDivisionError:
    print('Error is found: 0000X3445056jf05')
else:
    print('Else' +str(c))
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[]
try:
    c = a/b
    e=d[2]
except ZeroDivisionError:
    print('Error is found: 0000X3445056jf05')
else:
    print('Else' +str(c))
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[2,22,222]
try:
    c = a/b
    e=d[2]
except ZeroDivisionError:
    print('Error is found: 0000X3445056jf05')
except IndexError:
    print('Error found: 00003774Xsft2')
else:
    print(c)
    print(e)
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[2,22,222]
try:
    c = a/b
    e=d[5]
except ZeroDivisionError:
    print('Error is found: 0000X3445056jf05')
except IndexError:
    print('Error found: 00003774Xsft2')
else:
    print(c)
    print(e)
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[2,22,222]
try:
    c = a/b
    e=d[2]
except:
    print('Error found')
else:
    print(c)
    print(e)
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[2,22,222]
try:
    c = a/b
    e=d[2]
except Exception as e:
    print('Handle Error == ' +str(e))
else:
    print(c)
    print(e)
```

```python
a =int(input('What is your number? '))
b =int(input('What is your number? '))
d=[2,22,222]
try:
    c = a/b
    e=d[2]
except Exception as e:
    print('Handle Error == ' +str(e))
else:
    print(c)
    print(e)
finally:
    print('Finally will always run.')
```

# you can raise your own exception.

```python
class Students():
    def __init__(self,name,age):
        if age>80 or age<16:
             raise Exception('Age cannot be greater than 80 and less than 16')
        self.name = name
        self.age = age
```

```python
s1 = Students('Umer',17)
```
